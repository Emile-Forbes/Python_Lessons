##2.6 Python Sets
A set in Python is an unordered collection of unique items. This means:

It cannot contain duplicate elements.

The order of elements in a set is not preserved (i.e., the elements have no specific order).

Sets are used when you want to store a collection of items where duplicates are not allowed, and the order doesn't matter

```python
my_set = {1, 2, 3, 4}
print(my_set)  # Output: {1, 2, 3, 4}

# Duplicates are automatically removed
my_set_with_duplicates = {1, 2, 3, 3, 4}
print(my_set_with_duplicates)  # Output: {1, 2, 3, 4}

```

```python
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

# Union
print(set1 | set2)  # Output: {1, 2, 3, 4, 5, 6}

# Intersection
print(set1 & set2)  # Output: {3, 4}

# Difference
print(set1 - set2)  # Output: {1, 2}
```

```python
#playing with list and sets
# Define two lists
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]

# Convert lists to sets
set1 = set(list1)
set2 = set(list2)

# Apply Union (all unique elements from both sets)
union_result = set1 | set2  # or set1.union(set2)
print("Union:", union_result)

# Apply Intersection (common elements between both sets)
intersection_result = set1 & set2  # or set1.intersection(set2)
print("Intersection:", intersection_result)

# Apply Difference (elements in set1 but not in set2)
difference_result = set1 - set2  # or set1.difference(set2)
print("Difference (set1 - set2):", difference_result)

# Apply Symmetric Difference (elements in either set, but not in both)
symmetric_difference_result = set1 ^ set2  # or set1.symmetric_difference(set2)
print("Symmetric Difference:", symmetric_difference_result)
```

#3. Logical Operators


## Used to combine conditional statements.

| Operator | Description                          | Example            |
| -------- | ------------------------------------ | ------------------ |
| `and`    | Returns True if both are True        | `x > 5 and y < 10` |
| `or`     | Returns True if at least one is True | `x > 5 or y > 10`  |
| `not`    | Reverses the result                  | `not(x > 5)`       |


```python
x = 10
y = 4

print(x > 5 and y < 5)   # True
print(x < 5 or y < 5)    # True
print(not(x > 5))        # False
```

import os
import glob
import pandas as pd

def merge_csv_files(folder_path: str) -> pd.DataFrame:
    files = glob.glob(os.path.join(folder_path, "*.csv"))
    
    if not files:
        raise FileNotFoundError("No CSV files found in the specified folder.")
    
    df_list = []
    for file in files:
        try:
            df = pd.read_csv(file)
            df_list.append(df)
        except Exception as e:
            print(f"Error reading {file}: {e}")
    
    return pd.concat(df_list, ignore_index=True)


```python
#7 data transformation with decorators: Decorators are used to add extra fucntions without to a code without modifying it
import time

def timing_decorator(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        print(f"{func.__name__} took {time.time() - start:.2f}s")
        return result
    return wrapper

@timing_decorator
def transform_data(df):
    df['log_income'] = np.log1p(df['income'])
    return df

```

```python
import matplotlib.pyplot as plt

def plot_distribution(data, column, bins=10, color='blue'):
    """
    Plots a histogram to show the distribution of values in a specified column.

    Parameters:
    - data (DataFrame): The DataFrame containing the data.
    - column (str): The name of the column to visualize.
    - bins (int): Number of bins to use in the histogram.
    - color (str): Color of the bars in the histogram.

    Returns:
    - None
    """
    plt.figure(figsize=(10, 6))
    plt.hist(data[column], bins=bins, color=color, edgecolor='black', alpha=0.7)
    plt.title(f'Distribution of {column}', fontsize=14)
    plt.xlabel(column, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()


```

#10 Object Oriented Programming in python

OOP (Object-Oriented Programming) is a programming paradigm based on "objects", which bundle data (attributes) and behavior (methods).


| Concept           | Description                                      | Python Keyword |
| ----------------- | ------------------------------------------------ | -------------- |
| **Class**         | Blueprint for creating objects                   | `class`        |
| **Object**        | Instance of a class                              |               |
| **Attributes**    | Variables that hold data in a class              |               |
| **Methods**       | Functions defined inside a class                 | `def`          |
| **Constructor**   | Initializes the object                           | `__init__`     |
| **Inheritance**   | Subclassing to extend/override functionality     | `class B(A)`   |
| **Encapsulation** | Restricting access to internal variables/methods | `_` or `__`    |
| **Polymorphism**  | Overriding behavior with same interface          | Overriding     |

```python
#3 OOP in Data Engineering: Data Pipeline Class
import pandas as pd

class DataPipeline:
    def __init__(self, filepath):
        # self.filepath should be the string path to the file
        self.filepath = filepath
        self.data = None

    def load(self):
        # Use the stored string filepath to read the CSV
        self.data = pd.read_csv(self.filepath)

    def clean(self):
        self.data.dropna(inplace=True)
        self.data.columns = [col.lower() for col in self.data.columns]

    def transform(self, func):
        self.data = func(self.data)

    def save(self, path):
        self.data.to_csv(path, index=False)

# Assign the string filepath to the variable
filepath_string = "/content/gdrive/My Drive/Colab Notebooks/1 Data Cleansing/BL-Flickr-Images-Book.csv"

# Pass the string filepath to the DataPipeline constructor
data_pipeline = DataPipeline(filepath_string)

# Now, the load method will correctly read the CSV using the string path
data_pipeline.load()
data_pipeline.clean()
data_pipeline.save("/content/gdrive/My Drive/Colab Notebooks/22 Python Course/cleaned_data.csv")
```
